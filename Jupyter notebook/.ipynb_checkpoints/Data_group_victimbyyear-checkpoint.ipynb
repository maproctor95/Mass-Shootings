```python
import pandas as pd
```

```python
path = 'data/_fixed.csv'
mass_shooting = pd.read_csv(path)
```

```python
mass_shooting
```

```python
mass_shooting.info()
```

```python
mass_shooting['date'] = pd.to_datetime(mass_shooting['date'])
```

```python
group_year = mass_shooting.groupby(mass_shooting.date.dt.year)['total_number_of_victims'].sum()
group_year
```

```python
victim_year = pd.DataFrame(group_year, columns=['total_number_of_victims'])
victim_year
```

```python
victim_year.to_csv('Data/victim_year.csv', index = True)
```
